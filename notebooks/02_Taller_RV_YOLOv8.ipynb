```python
# --- Celda 3: Carga del Modelo de Segmentación ---
from ultralytics import YOLO

# Carga el modelo YOLOv8-nano pre-entrenado para SEGMENTACIÓN
model = YOLO('yolov8n-seg.pt')

print("¡Modelo YOLOv8-seg cargado con éxito!")
```

```python
# --- Celda 4: Ejecutar Inferencia ---

# 'source=' es la imagen que queremos analizar.
# 'save=True' le dice a YOLO que guarde la imagen con los resultados dibujados.
print("Ejecutando inferencia (detección + segmentación)...")
results = model(source='bus.jpg', save=True)

print("¡Inferencia completada! La imagen con los resultados se ha guardado.")
```

```python
# --- Celda 5: Mostrar Resultados ---
from IPython.display import Image

# Muestra la imagen que YOLO acaba de guardar
display(Image(filename='runs/segment/predict/bus.jpg'))
```
